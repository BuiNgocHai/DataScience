* Clean columns names ( lÃ m sáº¡ch tÃªn cÃ¡c cá»™t)

```python
def clean_label(s):
    s=s.strip()
    s=s.replace("Operating System", "os")
    s=s.replace(" ","_")
    s=s.replace("(","")
    s=s.replace(")","")
    s=s.lower()
    return s
    
```

```python
laptops.columns = [clean_label(label) for label in laptops.columns]
laptops.columns
```

- Convert RAM, Price, Weight into numerical columns (create new columns) (chuyá»ƒn Ä‘á»•i cÃ¡c giÃ¡ trá»‹ cá»™t RAM, Price, Weight sang giÃ¡ trá»‹ sá»‘)

* Chuyá»ƒn Ä‘á»•i cá»™t ram.

```python
laptops.ram.unique() #.unique tráº£ vá» giÃ¡ trá»‹ khÃ¡c nhau trÃªn cá»™t ram
```

```python
laptops['ram_gb'] = laptops.ram.str.slice(0,-2).astype(int) # .slice(0,-2) bá» hai pháº§n tá»­ cuá»‘i cá»§a máº£ng.
```

```python
laptops.ram_gb
```

* Chuyá»ƒn Ä‘á»•i cá»™t price_euros

```python
laptops['price_euros_nums'] = laptops.price_euros.str.replace(',','.').astype(float) # thay tháº¿ dáº¥u ',' sang '.' vÃ¬ kiá»ƒu sá»‘ k nháº­n dáº¥u pháº©y.
```

```python
laptops.price_euros_nums.unique()
```

* Chuyá»ƒn Ä‘á»•i cá»™t cÃ¢n náº·ng

```python
laptops['weight_kg'] = laptops.weight.str.slice(0,-2).astype(float)
```

 á»ž Ä‘Ã¢y chÃºng ta tháº¥y lá»—i khÃ´ng thá»ƒ chuyá»ƒn Ä‘á»•i sang kiá»ƒu float(could not convert string to float: '4k') nÃªn chÃºng ta sáº½ kiá»ƒm tra xem máº£ng cá»§a chÃºng ta sáº½ cÃ³ pháº§n tá»­ Ä‘áº·c biá»‡t nÃ o khÃ´ng

```python
laptops.weight[~laptops.weight.str.match(".*[0-9](kg)$")] 
# "." so khá»›p kÃ½ tá»±
# '.[0-9](kg)' tÃ¬m táº¥t cáº£ cÃ¡c string dáº¡ng sá»‘ tá»« 0 Ä‘áº¿n 9  vÃ  kg
# '*' kÃ½ tá»± nÃ y Ä‘áº¡i diá»‡n cho cÃ³ or k cÃ³ kÃ½ tá»± trÆ°á»›c nÃ³
# '$' dáº¥u káº¿t thÃºc
# '~' cá»© nghÄ© lÃ  Ä‘Ã¢y lÃ  dáº¥u khÃ¡c Ä‘i :v (tÃ¬m cÃ¡c pháº§n tá»­ khÃ¡c Ä‘á»‹nh dang .[0-9]kg)
```

```python
laptops['weight_kg'] = laptops.weight.str.replace("kg","").str.replace("s","").astype(float)
```

```python
laptops.weight_kg.unique()
```

- Extract CPU speed from CPU column

```python
laptops["cpu_speed"]=laptops.cpu.str.replace("GHz", "").str.extract(" ([0-9\.]*)$").astype("float")
```

```python
laptops.cpu_speed
```

2/ Convert Screen Size column in to numeric and use that column to answer:

- What is the popular screen size ?
- What is the minimum screen size of Ultrabook ?
- What is the average screen size of gaming laptop ?

```python
laptops["screen_size_inch"]=laptops.screen_size.str.slice(0,-1).astype(float)
laptops.screen_size_inch
```

```python
print("The popular screen size is ",laptops.screen_size.loc[laptops.screen_size.value_counts().max()])
```

```python
print("The minimum screen size of Ultrabook is ",laptops.screen_size.loc[laptops.category=="Ultrabook"].min())
```

```python
print("The average screen size of gaming laptop is",laptops.screen_size_inch.loc[laptops.category=="Gaming"].mean())
```

3/ Extract Screen Resolution (for example 2880x1800), and answer:

- How many laptop has full hd (1920x1080) screen ?
- What is the popular screen resolution of Macbook ?

```python
laptops["screen_resolution"]=laptops.screen.str.extract('(\d\d\d\d.\d\d\d*)',expand=True)
laptops.screen_resolution
```

```python
print("So laptop co man hinh Full HD 1920x1080 la: ",laptops.screen_resolution.loc[laptops.screen_resolution=="1920x1080"].count())
```

- Xá»­ lÃ½ model_name

```python
laptops.model_name.unique()
```

```python
def clean_model_name(s):
    s=s.strip()
    s=s.replace(" ","_")
    s=s.replace("(","")
    s=s.replace(")","")
    s=s.replace("-","_")
    s=s.replace("/","_")
    s=s.lower()
    return s
```

```python
laptops.model_name= [clean_model_name(lable_name) for lable_name in laptops.model_name]
laptops.model_name
```
